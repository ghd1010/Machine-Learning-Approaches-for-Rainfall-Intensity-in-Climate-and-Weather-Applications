```python
# importing libraries and data
import numpy as np
import os
import pandas as pd
#read the path
file_path=("/content/uganda")
#list all the files from the directory
file_list = os.listdir(file_path)
file_list
```

```python
file_list = os.listdir(file_path)

dfs = [] # collect the files in an array

for file in file_list:
    file_full_path = os.path.join(file_path, file)
    df_temp = pd.read_csv(file_full_path)
    dfs.append(df_temp)

df_append = pd.concat(dfs, ignore_index=True)
df_append.head(5)
```

```python
# shuffle data
df_append_shuffled = df_append.sample(frac=1, random_state=42)
df_append_shuffled.head(5)
```
